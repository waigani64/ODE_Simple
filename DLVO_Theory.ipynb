```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k_b = 1.38e-23  # Boltzmann constant (J/K)
eps_0 = 8.85e-12  # Permittivity of free space (F/m)
e = 1.602e-19  # Elementary charge (C)
A = 1e-19  # Hamaker constant (J)
kappa = 1e-9  # Debye length (m)

# Distance range
r = np.linspace(1e-9, 10e-9, 1000)  # Distance between particles (m)

# DLVO theory calculations
# Electrostatic repulsion
Z = e**2 / (4 * np.pi * eps_0 * k_b * 298.15 * kappa * r)
F_elec = Z * 1e9  # Convert to nanonewtons (nN)
# Van der Waals attraction
F_vdw = - A / 6 * (2 / (r + 2*kappa)**2 - 2 / (r + kappa)**2 + np.log((r + 2*kappa)/(r + kappa)))
# Total force
F_total = F_elec + F_vdw

# Plot the results
plt.plot(r*1e9, F_elec, label="Electrostatic repulsion")
plt.plot(r*1e9, F_vdw, label="Van der Waals attraction")
plt.plot(r*1e9, F_total, label="Total force")
plt.xlabel("Distance (nm)")
plt.ylabel("Force (nN)")
plt.legend()
plt.show()
```
